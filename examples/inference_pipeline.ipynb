```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from blindspot_denoise import pipeline
```

```python
MODEL_PATH = Path("checkpoints/model.pt")  # example path
DATA_PATH = Path("data/example.npy")      # example path (npy file)

# Load input as in-memory numpy array
data = np.load(DATA_PATH)  # (N, T, X) or (T, X)
if data.ndim == 2:
    data = data[None, ...]
N, T, X = data.shape
```

```python
# Create inference pipeline
bs = 8
pipe = pipeline(model=MODEL_PATH, device="auto", batch_size=bs)

# Loop over input in batches and collect outputs
outs = np.empty_like(data)
for i in range(0, N, bs):
    batch = data[i:i+bs]
    outs[i:i+bs] = pipe(batch)
```

```python
# Plot before/after for the first sample
idx = 0
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(data[idx], aspect='auto', cmap='gray')
axes[0].set_title('Input')
axes[1].imshow(outs[idx], aspect='auto', cmap='gray')
axes[1].set_title('Denoised')
plt.tight_layout()
plt.show()
```
